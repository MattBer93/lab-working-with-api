```python
headers = {
	"X-RapidAPI-Key": "0eda373a4cmsh180112ca5fb2dcep162b5djsn68f33e620891",
	"X-RapidAPI-Host": "aerodatabox.p.rapidapi.com"
}

def get_departures_table():
    location = input('Where are you leaving from? Please enter the exact icao code: ').upper()
    timeframe = input('When? Please add the time in this format:\n \"YYYY-MM-DDThh:mm/YYYY-MM-DDThh:mm\"')
    carrier = input('Which company do you want to fly with?').upper()

    url = f"https://aerodatabox.p.rapidapi.com/flights/airports/icao/{location}/{timeframe}"
    querystring = {"direction":"Departure","withCancelled":"false","withCodeshared":"true","withLocation":"false"}

    response = requests.get(url, headers=headers, params=querystring)
    resp_json = response.json()

    results_list = []
    for i in range(0, len(resp_json['departures'])):
        if carrier == resp_json['departures'][i]['airline']['name']:
            results_list.append(resp_json['departures'][i])


    json_list = json.dumps(results_list)
    return pd.json_normalize(results_list).T
```
